<h2>text to pandas dataframe data conversion using regular expression</h2>

```python
f = open('chat_of_3log.txt', 'r', encoding='utf-8')
data = f.read()
print(data)
```

```python
pattern = '\d{1,2}/\d{1,2}/\d{2},\s\d{1,2}:\d{2}\s[ap]m?\s*-\s'

msg = re.split(pattern,data)[1:]
print(msg)
```

```python
dates = re.findall(pattern,data)
dates
```

```python
# if len(dates) != len(msg):
#     # If lengths(no. of rows) are unequal, adjust the lists to match
#     min_len = min(len(dates), len(msg))
#     dates = dates[:min_len]
#     msg = msg[:min_len]
```

```python
df = pd.DataFrame({'date': dates, 'msg': msg})
df['date'] = df['date'].str.replace('\u202f', ' ', regex=False)
df['date'] = df['date'].str.replace('\xa0', ' ', regex=False)  # optional extra safety
df['date'] = df['date'].str.strip() 

df['date'] = pd.to_datetime(df['date'], format='%d/%m/%y, %I:%M %p -')
df.head(5)
```

```python
users = []
messages_list = []  # Changed variable name to avoid conflict

for msg in df['msg']:  # Changed loop variable to 'msg'
    entry = re.split('([\w\W]+?):\s', msg)
    if len(entry) >= 3:  # Check if split found a colon (returns 3 parts)
        users.append(entry[1])  # User is in the second group
        messages_list.append(entry[2])  # Message is in the third group
    else:
        # Handle cases where the message doesn't have a colon
        users.append('Group_Notification')
        messages_list.append(msg)  # Use the original message

df['user'] = users
df['message'] = messages_list
df.drop(columns=['msg'], inplace=True)
```

<h1>main

```python
import re
import pandas as pd
from datetime import datetime

f = open('chat_of_3log.txt', 'r', encoding='utf-8')
data = f.read()

# Refined pattern
pattern = r"(\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s[ap]m)\s-\s([^:]+):\s(.+)"

# Find all matches
matches = re.findall(pattern, data)

# Extract datetime and message
dates = [match[0] for match in matches]
user = [match[1] for match in matches]
messages = [match[2] for match in matches]

# Convert to datetime
dates = pd.to_datetime(dates, format='%d/%m/%y, %I:%M %p')  # Adjust format if needed

# Create DataFrame
df = pd.DataFrame({
    'datetime': dates,
    'user': user,
    'message': messages
})

df.sample(10)
```
